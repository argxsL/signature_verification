```python
import os
import cv2
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
IMG_SIZE = 100

def preprocess(img_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    return img.flatten()
```

```python
id = "050"
def load_049(base_path):
    X, y = [], []

    real_path = os.path.join(base_path, id)
    fake_path = os.path.join(base_path, id+"_forg")

    for file in os.listdir(real_path):
        X.append(preprocess(os.path.join(real_path, file)))
        y.append(1)   # real

    for file in os.listdir(fake_path):
        X.append(preprocess(os.path.join(fake_path, file)))
        y.append(0)   # fake

    return np.array(X), np.array(y)
```

```python
TRAIN_PATH = "/content/drive/MyDrive/signatures/train"
TEST_PATH  = "/content/drive/MyDrive/signatures/test"

X_train, y_train = load_049(TRAIN_PATH)
X_test, y_test   = load_049(TEST_PATH)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
```

```python
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)

# Show first few predictions
for i in range(5):
    print("Predicted:", "Real" if y_pred[i] == 1 else "Fake",
          "| Actual:", "Real" if y_test[i] == 1 else "Fake")
```

```python
acc = accuracy_score(y_test, y_pred)
print("Accuracy for ID 049:", acc)
```
